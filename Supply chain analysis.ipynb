#Supply chain analysis

```python
#Import Laibraries

import pandas as pd
import numpy as np
```

```python
# Importing & Cleaning Loads Table 

loads_df=pd.read_csv('loads.csv')
loads_df.info()
loads_df['load_date']=pd.to_datetime(loads_df['load_date'])
loads_df.columns = loads_df.columns.str.strip()
loads_df.info()
loads_df.head()
```

```python
# Importing & Cleaning customers Table
 
customers_df=pd.read_csv('customers.csv')
customers_df.info()
customers_df['contract_start_date']=pd.to_datetime(customers_df['contract_start_date'])
customers_df.columns = customers_df.columns.str.strip()
customers_df.info()
customers_df.head()
```

```python
# Importing & Cleaning drivers Table

drivers_df=pd.read_csv('drivers.csv')
drivers_df.info()
drivers_df['termination_date']=pd.to_datetime(drivers_df['termination_date'])
drivers_df['termination_date'] = drivers_df['termination_date'].fillna('On_Duty')
drivers_df['first_name'] = drivers_df['first_name'].str.title()
drivers_df['last_name'] = drivers_df['last_name'].str.title()
drivers_df.columns = drivers_df.columns.str.strip()
drivers_df.info()
drivers_df.head()
```

```python
# Importing & Cleaning delivery_events Table

delivery_events_df=pd.read_csv('delivery_events.csv')
delivery_events_df.info()
delivery_events_df['actual_datetime']=pd.to_datetime(delivery_events_df['actual_datetime'])
delivery_events_df['scheduled_datetime']=pd.to_datetime(delivery_events_df['scheduled_datetime'])
delivery_events_df['location_city'] = delivery_events_df['location_city'].str.title()
delivery_events_df.columns = delivery_events_df.columns.str.strip()
delivery_events_df.info()
delivery_events_df.head()
```

```python
# Importing & Cleaning driver_monthly_metrics Table

driver_monthly_metrics_df=pd.read_csv('driver_monthly_metrics.csv')
driver_monthly_metrics_df.info()
driver_monthly_metrics_df['month']=pd.to_datetime(driver_monthly_metrics_df['month'])
drivers_df.columns = drivers_df.columns.str.strip()
driver_monthly_metrics_df.info()
driver_monthly_metrics_df.head()
```

```python
# Importing & Cleaning facilities Table

facilities_df=pd.read_csv('facilities.csv')
facilities_df['city'] = facilities_df['city'].str.title()
facilities_df['facility_name'] = facilities_df['facility_name'].str.title()
facilities_df['facility_type'] = facilities_df['facility_type'].str.title()
facilities_df.columns = facilities_df.columns.str.strip()
facilities_df.info()
facilities_df.head()
```

```python
# Importing & Cleaning fuel_purchases Table

fuel_purchases_df=pd.read_csv('fuel_purchases.csv')
fuel_purchases_df.info()
fuel_purchases_df=fuel_purchases_df.dropna()
fuel_purchases_df['purchase_date']=pd.to_datetime(fuel_purchases_df['purchase_date'])
fuel_purchases_df['location_city'] = fuel_purchases_df['location_city'].str.title()
fuel_purchases_df.columns = fuel_purchases_df.columns.str.strip()
fuel_purchases_df.info()
fuel_purchases_df.head()
```

```python
# Importing & Cleaning maintenance_records Table

maintenance_records_df=pd.read_csv('maintenance_records.csv')
maintenance_records_df.info()
maintenance_records_df['maintenance_date']=pd.to_datetime(maintenance_records_df['maintenance_date'])
maintenance_records_df['facility_location'] = maintenance_records_df['facility_location'].str.title()
maintenance_records_df.columns = maintenance_records_df.columns.str.strip()
maintenance_records_df.info()
maintenance_records_df.head()
```

```python
# Importing & Cleaning routes Table

routes_df=pd.read_csv('routes.csv')
routes_df['destination_city'] = routes_df['destination_city'].str.title()
routes_df.columns = routes_df.columns.str.strip()
routes_df.info()
routes_df.head()
```

```python
# Importing & Cleaning safety_incidents Table

safety_incidents_df=pd.read_csv('safety_incidents.csv')
safety_incidents_df.info()
safety_incidents_df=safety_incidents_df.dropna(subset=["truck_id","driver_id"])
safety_incidents_df['incident_date']=pd.to_datetime(safety_incidents_df['incident_date'])
safety_incidents_df['location_city'] = safety_incidents_df['location_city'].str.title()
safety_incidents_df.columns = safety_incidents_df.columns.str.strip()
safety_incidents_df.info()
safety_incidents_df.head()
```

```python
# Importing & Cleaning trailers Table

trailers_df=pd.read_csv('trailers.csv')
trailers_df.info()
trailers_df['acquisition_date']=pd.to_datetime(trailers_df['acquisition_date'])
trailers_df.columns = trailers_df.columns.str.strip()
trailers_df.info()
trailers_df.head()
```

```python
# Importing & Cleaning trips Table

trips_df=pd.read_csv('trips.csv')
trips_df.info()
trips_df['dispatch_date']=pd.to_datetime(trips_df['dispatch_date'])
trips_df=trips_df.dropna(subset=["driver_id","truck_id","trailer_id"])
trips_df.columns = trips_df.columns.str.strip()
trips_df.info()
trips_df.head()
```
