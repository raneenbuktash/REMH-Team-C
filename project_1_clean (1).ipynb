هنزامن الجداول باستخدام trip id and compare each colomn

```python
# 1. تجهيز القوائم للربط (Sets are faster for searching)
valid_trips = set(all_data['trips']['trip_id'])
valid_loads = set(all_data['trips']['load_id'])

print(" بدء عملية المزامنة لضمان ترابط البيانات (Referential Integrity):")
print("-" * 50)#مسافة بخط

# 2. كود للمزامنة التلقائية لأي جدول مرتبط بالرحلة مباشرة
for name, df in all_data.items():
    # تخطي جدول الرحلات نفسه لأنه المرجع
    if name == 'trips': continue

    # المزامنة عبر trip_id
    if 'trip_id' in df.columns:
        initial_count = len(df)
        all_data[name] = df[df['trip_id'].isin(valid_trips)]
        print(f"✅ تم مزامنة {name:20} | حذف: {initial_count - len(all_data[name]):<5} صفوف")

# 3. مزامنة جدول الأحمال (حالة خاصة لأنه يرتبط عبر load_id)
if 'loads' in all_data:
    initial_count = len(all_data['loads'])
    all_data['loads'] = all_data['loads'][all_data['loads']['load_id'].isin(valid_loads)]
    print(f"✅ تم مزامنة {'loads':20} | حذف: {initial_count - len(all_data['loads']):<5} صفوف (عبر load_id)")

print("-" * 50)
print("🎯 انتهت المزامنة.. الآن كل الجداول مرتبطة برحلات حقيقية وسليمة.")
```

```python
# 1. تجهيز القوائم للربط (Sets) لسرعة البحث
valid_trips = set(all_data['trips']['trip_id'])
valid_loads = set(all_data['trips']['load_id'])

print(" تقرير مزامنة البيانات (Data Synchronization Report):")
print("-" * 65)
print(f"{'الجدول':<20} | {'قبل':<8} | {'بعد':<8} | {'المحذوف':<8}")
print("-" * 65)

# 2. حلقة المزامنة التلقائية للجداول المرتبطة بـ trip_id
for name, df in all_data.items():
    if name == 'trips': continue

    if 'trip_id' in df.columns:
        before_count = len(df)
        all_data[name] = df[df['trip_id'].isin(valid_trips)]
        after_count = len(all_data[name])
        deleted = before_count - after_count

        print(f"{name:<20} | {before_count:<8} | {after_count:<8} | {deleted:<8}")

# 3. مزامنة جدول الأحمال (Loads) حالة خاصة عبر load_id
if 'loads' in all_data:
    before_count = len(all_data['loads'])
    all_data['loads'] = all_data['loads'][all_data['loads']['load_id'].isin(valid_loads)]
    after_count = len(all_data['loads'])
    deleted = before_count - after_count

    print(f"{'loads (via load_id)':<20} | {before_count:<8} | {after_count:<8} | {deleted:<8}")

print("-" * 65)
print(" تم تنقية جميع الجداول من البيانات غير المرتبطة برحلات سليمة.")
```

```python
# 1. أولاً: نحدد المراجع السليمة (الرحلات والأحمال التي لم تُحذف)
valid_trips = set(all_data['trips']['trip_id'])
valid_loads = set(all_data['trips']['load_id'])

print(" إعادة مزامنة البيانات (Data Synchronization Report):")
print("-" * 70)
# رأس الجدول للتنسيق
print(f"{'الجدول':<25} | {'قبل الحذف':<10} | {'بعد الحذف':<10} | {'المحذوف':<10}")
print("-" * 70)

# 2. حلقة تمر على كل الجداول لمزامنة أي جدول يحتوي على trip_id
for name, df in all_data.items():
    if name == 'trips': continue  # نتخطى جدول الرحلات لأنه هو المرجع الأساسي

    if 'trip_id' in df.columns:
        before_count = len(df)
        # الاحتفاظ فقط بالصفوف المرتبطة برحلات سليمة
        all_data[name] = df[df['trip_id'].isin(valid_trips)].copy()
        after_count = len(all_data[name])
        deleted = before_count - after_count

        print(f"{name:<25} | {before_count:<10} | {after_count:<10} | {deleted:<10}")

# 3. مزامنة جدول الأحمال (حالة خاصة لأنه يرتبط بـ load_id)
if 'loads' in all_data:
    before_count = len(all_data['loads'])
    all_data['loads'] = all_data['loads'][all_data['loads']['load_id'].isin(valid_loads)].copy()
    after_count = len(all_data['loads'])
    deleted = before_count - after_count

    print(f"{'loads (via load_id)':<25} | {before_count:<10} | {after_count:<10} | {deleted:<10}")

print("-" * 70)
print(" تمت إعادة المزامنة بنجاح! الآن جميع الجداول متوافقة مع الـ ERD.")
```

```python
# --- تنظيف جدول السائقين (Drivers Table Cleaning) ---

# 1. تحويل التواريخ من نصوص إلى Datetime
date_cols = ['hire_date', 'date_of_birth']
for col in date_cols:
    all_data['drivers'][col] = pd.to_datetime(all_data['drivers'][col])

# 2. معالجة حالة السائقين النشطين
all_data['drivers']['termination_date'] = all_data['drivers']['termination_date'].fillna('Active')

# 3. توحيد نصوص الأسماء والولايات (Standardization)
all_data['drivers']['first_name'] = all_data['drivers']['first_name'].str.strip().str.title()
all_data['drivers']['last_name'] = all_data['drivers']['last_name'].str.strip().str.title()
all_data['drivers']['license_state'] = all_data['drivers']['license_state'].str.strip().str.upper()

# 4. تحويل الأعمدة الى category لتحسين الأداء (Categorical Data)
cat_cols = ['employment_status', 'license_state', 'cdl_class']
for col in cat_cols:
    all_data['drivers'][col] = all_data['drivers'][col].astype('category')

print("✅ تم تنظيف جدول السائقين بالكامل وتصنيف البيانات.")
print("-" * 30)
print(all_data['drivers'].info())
```

```python
# إضافة عمود يوضح الحالة بناءً على الرقم
def get_status(delay):
    if delay > 0:
        return 'Late'
    elif delay < 0:
        return 'Early'
    else:
        return 'Exactly on Time'

all_data['delivery_events']['arrival_status'] = all_data['delivery_events']['arrival_delay_minutes'].apply(get_status)

# عرض العينة مرة أخرى لرؤية الفرق
display(all_data['delivery_events'][['event_type', 'arrival_delay_minutes', 'arrival_status', 'on_time_flag']].head())
```

```python
import pandas as pd
import os

# 1. تحديد المسار الأساسي
base_path = '/content/drive/MyDrive/Data Analysis DEPI Project/Depi Final Project R/'

# 2. قائمة بالملفات التي نريد فحص أعمدتها والـ Nulls فيها
files_to_audit = {
    'Fuel': 'fuel_purchases.csv',
    'Trips': 'trips.csv',
    'Loads': 'loads.csv'
}

print("🔍 تقرير فحص جودة البيانات (Data Quality Audit):\n")

for label, file_name in files_to_check.items():
    file_path = os.path.join(base_path, file_name)

    if os.path.exists(file_path):
        df = pd.read_csv(file_path)

        print(f"📋 جدول [{label}]:")
        print(f"   - أسماء الأعمدة: {list(df.columns)}")

        # فحص الـ Nulls في هذا الجدول
        nulls = df.isnull().sum()
        actual_nulls = nulls[nulls > 0]

        if not actual_nulls.empty:
            print(f"   - ⚠️ تم العثور على قيم مفقودة (Nulls):")
            for col, count in actual_nulls.items():
                percentage = (count / len(df)) * 100
                print(f"     * عمود [{col}]: {count} قيمة فارغة ({percentage:.2f}%)")
        else:
            print("   - ✅ لا توجد قيم مفقودة في هذا الجدول.")
        print("-" * 50)
    else:
        print(f"❌ الملف {file_name} غير موجود في المسار المحدد.")
```

```python
import pandas as pd
import os

# 1. المسار الأساسي (تأكدي من وجود الشرطة المائلة في النهاية)
base_path = '/content/drive/MyDrive/Data Analysis DEPI Project/Depi Final Project R/'

# 2. الحصول على قائمة بكل ملفات الـ CSV في المجلد
csv_files = [f for f in os.listdir(base_path) if f.endswith('.csv')]

# 3. إنشاء قائمة لتخزين النتائج
audit_results = []

print(f"🚀 بدء فحص {len(csv_files)} ملفاً في المجلد...\n")

for file_name in csv_files:
    file_path = os.path.join(base_path, file_name)
    df = pd.read_csv(file_path)

    # حساب الإحصائيات لكل ملف
    total_rows = len(df)
    null_counts = df.isnull().sum().sum() # إجمالي الـ Nulls في الجدول كله
    cols_with_nulls = df.columns[df.isnull().any()].tolist() # الأعمدة المصابة

    audit_results.append({
        'Table Name': file_name,
        'Total Rows': total_rows,
        'Total Columns': len(df.columns),
        'Total Nulls': null_counts,
        'Null Percentage': (null_counts / (total_rows * len(df.columns))) * 100,
        'Affected Columns': cols_with_nulls
    })

# 4. تحويل النتائج لجدول منظم
audit_df = pd.DataFrame(audit_results)

# 5. عرض النتائج مرتبة حسب الأكثر احتواءً على Nulls
print("📊 تقرير جودة البيانات الشامل:")
display(audit_df.sort_values(by='Total Nulls', ascending=False))
```

```python
import pandas as pd

# 1. تحميل الملف
fuel_df = pd.read_csv('/content/drive/MyDrive/Data Analysis DEPI Project/Depi Final Project R/fuel_purchases.csv')

# 2. حذف أي صفوف فارغة تماماً قد تكون في نهاية الملف
fuel_df = fuel_df.dropna(how='all')

# 3. التأكد من أن الأعمدة الحسابية أرقام فعلاً (أي نص سيتحول لـ Null ثم نعوضه)
cols_to_fix = ['gallons', 'price_per_gallon', 'total_cost']
for col in cols_to_fix:
    fuel_df[col] = pd.to_numeric(fuel_df[col], errors='coerce') # تحويل أي نص خطأ لـ Null
    # ملء أي Null ظهر بالمتوسط لضمان عدم رفض SQL
    fuel_df[col] = fuel_df[col].fillna(fuel_df[col].mean())

# 4. التعامل مع الـ IDs المفقودة (3988 قيمة)
# بما أن SQL يرفض الـ Null في الـ IDs، سنضع قيمة افتراضية مثل 'UNKNOWN'
fuel_df['driver_id'] = fuel_df['driver_id'].fillna('UNKNOWN')
fuel_df['truck_id'] = fuel_df['truck_id'].fillna('UNKNOWN')

# 5. حفظ الملف بصيغة نظيفة جداً وبدون فهرس
output_path = '/content/drive/MyDrive/Data Analysis DEPI Project/Depi Final Project R/fuel_purchases_v3_ready.csv'
fuel_df.to_csv(output_path, index=False)

print(f"✅ تم تجهيز الملف النهائي: {output_path}")
```

```python
import pandas as pd

# 1. تحميل ملف الوقود
file_path = '/content/drive/MyDrive/Data Analysis DEPI Project/Depi Final Project R/fuel_purchases.csv'
fuel_df = pd.read_csv(file_path)

# 2. حذف الصفوف التي تحتوي على Null في driver_id أو truck_id
# سنستخدم dropna ونحدد الأعمدة المستهدفة (subset)
fuel_df_cleaned = fuel_df.dropna(subset=['driver_id', 'truck_id']).copy()

# 3. التأكد من حذف أي صفوف وهمية فارغة تماماً (Ghost Rows)
fuel_df_cleaned = fuel_df_cleaned.dropna(how='all')

# 4. معالجة الأعمدة الحسابية (للتأكد من عدم وجود رموز غريبة تسبب رفض SQL)
cols_to_check = ['gallons', 'price_per_gallon', 'total_cost']
for col in cols_to_check:
    # تحويل العمود لرقم، وأي قيمة غير صالحة تتحول لـ NaN ثم تحذف
    fuel_df_cleaned[col] = pd.to_numeric(fuel_df_cleaned[col], errors='coerce')

# 5. حذف أي صفوف نتجت عن خطأ في التحويل الرقمي في الخطوة السابقة
fuel_df_cleaned = fuel_df_cleaned.dropna(subset=cols_to_check)

# 6. فحص نهائي للتأكد من نظافة البيانات
print(f"✅ عدد الصفوف قبل الحذف: {len(fuel_df)}")
print(f"✅ عدد الصفوف بعد الحذف: {len(fuel_df_cleaned)}")
print(f"✅ عدد القيم المفقودة الآن: {fuel_df_cleaned.isnull().sum().sum()}")

# 7. حفظ الملف بصيغة CSV نظيفة جداً
output_path = '/content/drive/MyDrive/Data Analysis DEPI Project/Depi Final Project R/fuel_purchases_final_v4.csv'
fuel_df_cleaned.to_csv(output_path, index=False)

print(f"\n🚀 الملف جاهز للرفع الآن: {output_path}")
```
